```python
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
def show(X):
    plt = matplotlib.pyplot
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot()
    plt.scatter(X[:, 0], X[:, 1], color="blue")
   
    plt.grid()
    plt.show()
```

```python
def show_k_means(X, y_km, km, clusters_num):
    plt = matplotlib.pyplot
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot()
    
    plt.title('K - Means')
    
    for i in range(0, clusters_num):
        plt.scatter( X[y_km == i, 0], X[y_km == i, 1], s=50, label=('cluster : '+str(i+1)))
        
    
    plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
    s=250, marker='*',
    c='white', edgecolor='black',
    label='centroids'
    )
    
    plt.legend(scatterpoints=1)
    plt.grid()
    plt.show()
        
        
```

```python
def show_dbscan(X, y_db):
    plt.figure(figsize=(10,7))
    plt.title('DBSCAN')
    plt.scatter(X[:,0], X[:,1],c=y_db, cmap='Paired')
    plt.grid()
    plt.show()
```

```python
def f_par_down(x):
    return -x**2
```

```python
def f_par_up(x):
    return ((x-10)**2)-150
```

```python
def f_lin_1(x):
    return 2*x+40
```

```python
def f_lin_2(x):
    return 2*x-40
```

```python
# Case 1
```

```python
# generate the data
x, y = [], []

n_dots = 200

tmp_x = np.random.randint(0, 2, n_dots) + np.random.normal(0, 2, n_dots)
tmp_y = np.random.randint(0, 2, n_dots) + np.random.normal(0, 2, n_dots)
x = np.append(x, tmp_x)
y = np.append(y, tmp_y)

tmp_x = np.random.randint(0, 2, n_dots) + np.random.normal(0, 2, n_dots)
tmp_y = np.random.randint(15, 20, n_dots) + np.random.normal(0, 2, n_dots)
x = np.append(x, tmp_x)
y = np.append(y, tmp_y)

tmp_x = np.random.randint(15, 30, n_dots) - np.random.normal(0, 2, n_dots)
tmp_y = np.random.randint(2, 5, n_dots) - np.random.normal(0, 2, n_dots)
x = np.append(x, tmp_x)
y = np.append(y, tmp_y)


tmp_x = np.random.randint(30, 50, n_dots) + np.random.normal(0, 3, n_dots)
tmp_y = np.random.randint(30, 50, n_dots) + np.random.normal(0, 3, n_dots)
x = np.append(x, tmp_x)
y = np.append(y, tmp_y)


x = x.reshape([-1, 1])
y = y.reshape([-1, 1])
X = np.hstack([x, y])

# illustrations
print("X shape : ", X.shape)
show(X)
```

```python
# KMeans algorythm
clusters_num = 4

# initialize KMeans model
km = KMeans(n_clusters=clusters_num, init='random')
# model training
y_km = km.fit_predict(X)
```

```python
# DBSCAN algorythm

# initialize DBSCAN model
db_scan = DBSCAN(eps=3, min_samples=5)
# model training
y_db = db_scan.fit_predict(X)
```

```python
# result KMeans
show_k_means(X, y_km, km, clusters_num)

# result DBSCAN
show_dbscan(X, y_db)
```
